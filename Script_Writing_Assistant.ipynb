```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load GPT-2 model and tokenizer
model_name = "gpt2"  # For simplicity, we use the smaller GPT-2 model

model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

model.eval()
```

```python
def generate_script(prompt, max_length=300, genre="tech", audience="Indian Bangalore"):
    # Customize the prompt for different genres
    prompt = f"Generate a {genre} YouTube script for an audience in {audience}. Topic: {prompt}"

    # Encode the input prompt
    inputs = tokenizer.encode(prompt, return_tensors="pt")

    # Generate the script
    outputs = model.generate(inputs, max_length=max_length, num_return_sequences=1, no_repeat_ngram_size=2, temperature=0.7)

    # Decode the generated script
    generated_script = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return generated_script
```

```python
import ipywidgets as widgets
from IPython.display import display

# Create widgets for genre and topic input
genre_widget = widgets.Dropdown(
    options=['Tech', 'Travel', 'Food', 'Lifestyle', 'Vlog'],
    value='Tech',
    description='Genre:',
)

topic_widget = widgets.Text(
    value='How to start a YouTube channel',
    description='Topic:',
)

generate_button = widgets.Button(description="Generate Script")

# Define the button click action
def on_button_click(b):
    genre = genre_widget.value
    topic = topic_widget.value
    script = generate_script(topic, genre=genre)
    print(f"Generated Script for {genre} - {topic}:")
    print(script)

generate_button.on_click(on_button_click)

# Display the widgets
display(genre_widget, topic_widget, generate_button)
```

```python
import os

# Function to generate script
def generate_script(prompt, genre="comedy", max_length=300, audience="Indian Bangalore"):
    # Customize the prompt for the specified genre and audience
    prompt = f"Generate a {genre} YouTube script for an audience in {audience}. Topic: {prompt}"

    # Encode the input prompt
    inputs = tokenizer.encode(prompt, return_tensors="pt")

    # Generate the script
    outputs = model.generate(inputs, max_length=max_length, num_return_sequences=1, no_repeat_ngram_size=2, temperature=0.7)

    # Decode the generated script
    generated_script = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return generated_script

# Function to save the script to a file with a dynamic name
def save_script_to_file(script, topic, genre):
    # Create a valid file name by removing any spaces and special characters
    file_name = f"{genre}_{topic.replace(' ', '_').replace('/', '_')}.txt"

    # Specify the path to save the file
    file_path = os.path.join("/content", file_name)

    # Save the script to the file
    with open(file_path, "w") as file:
        file.write(script)

    return file_path

# Example usage for Jordindian-style script generation
topic = "Funny experience in Bangalore traffic"
genre = "Comedy"
script = generate_script(topic, genre=genre)

# Save the generated script to a file with a dynamic name
file_path = save_script_to_file(script, topic, genre)

# Display the path where the file is saved
print(f"Script saved as: {file_path}")
```
